```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# D liu im trc v sau can thip
truoc = [68, 75, 80, 71, 90, 74, 85, 69, 77, 73]
sau   = [70, 78, 82, 72, 91, 76, 87, 70, 79, 74]

# To mt DataFrame cho d liu
data = pd.DataFrame({
    'truoc': truoc,
    'sau': sau
})

# To biu  Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Biu  Boxplot: im trc v sau can thip')
plt.ylabel('im s')
plt.xticks([0, 1], ['Trc', 'Sau'])
plt.show()

# To biu  Lineplot
plt.figure(figsize=(8, 6))
plt.plot(data.index, truoc, label='Trc', marker='o')
plt.plot(data.index, sau, label='Sau', marker='o')
plt.title('Biu  Lineplot: So snh im trc v sau can thip')
plt.xlabel('C nhn')
plt.ylabel('im s')
plt.legend()
plt.grid(True)
plt.show()
```
