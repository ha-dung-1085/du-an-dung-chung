```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Dữ liệu điểm trước và sau can thiệp
truoc = [68, 75, 80, 71, 90, 74, 85, 69, 77, 73]
sau   = [70, 78, 82, 72, 91, 76, 87, 70, 79, 74]

# Tạo một DataFrame cho dữ liệu
data = pd.DataFrame({
    'truoc': truoc,
    'sau': sau
})

# Tạo biểu đồ Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Biểu đồ Boxplot: Điểm trước và sau can thiệp')
plt.ylabel('Điểm số')
plt.xticks([0, 1], ['Trước', 'Sau'])
plt.show()

# Tạo biểu đồ Lineplot
plt.figure(figsize=(8, 6))
plt.plot(data.index, truoc, label='Trước', marker='o')
plt.plot(data.index, sau, label='Sau', marker='o')
plt.title('Biểu đồ Lineplot: So sánh điểm trước và sau can thiệp')
plt.xlabel('Cá nhân')
plt.ylabel('Điểm số')
plt.legend()
plt.grid(True)
plt.show()
```
